```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(-10,10,100)
y=np.sin(x)
z=np.cos(x)
```

```python
#plotting sine graph
plt.plot(x,y)
plt.xlabel('Angle in radians')
plt.ylabel('Value')
plt.title('Sine Wave graph')
plt.show()
```

```python
#plotting cosine graph
plt.plot(x,z)
plt.xlabel('Angle in radians')
plt.ylabel('Value')
plt.title('Cosine Wave Graph')
plt.show()
```

```python
#plotting a parabola
x=np.linspace(-10,10,1000)
y=np.square(x)
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('Y-axis')
plt.title('Parabola graph')
plt.show()
```

```python
#plotting a parabola
x=np.linspace(-10,10,50)
y=np.tan(x)
plt.plot(x,y)
plt.xlabel('x-axis')
plt.ylabel('Y-axis')
plt.title('Parabola graph')
plt.show()
```

```python
#plotting with points instead of curves
x=np.linspace(-10,10,50)
y=x**2
z=-y
plt.plot(x,y,'r+')
plt.plot(x,z,'g+')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Parabola')
plt.show()
```

# Scatter plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-3*np.pi,3*np.pi,200)
y=np.sin(x)
plt.scatter(x,y)
plt.show()
```

```python
x1=np.linspace(1,10,100)
y1=x1**2
x2=x1
y2=x1**3
plt.plot(x1,y1,'g',linewidth=3)
plt.plot(x2,y2,'r',linewidth=4)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Comaparison Graph')
plt.annotate(xy=[4,150],s='Y=x^3')
plt.annotate(xy=[7,80],s='y=x^2')
plt.show()
```

#Vertical Bar graph

```python
x1=np.linspace(1,10,10)
y1=x1**2
x2=x1
y2=x1**3
plt.bar(x2,y2,color='g')
plt.bar(x1,y1,color='r')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('verticla Bar Graph')
plt.legend(['x^3','x^2'])
plt.show()
```

#Horizontal Bar Graph

```python
x1=np.linspace(1,10,10)
y1=x1**2
x2=x1
y2=x1**3
plt.barh(x2,y2,color='g')
plt.barh(x1,y1,color='r')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('verticla Bar Graph')
plt.legend(['x^3','x^2'])
plt.show()
```

```python
x=[1,2,3,4,5]
y=[10,5,8,15,10]
plt.bar(x,y,color='r')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('horizontal bar graph')
plt.show()
```

```python
x=[1,2,3,4,5]
y=[10,5,8,15,10]
plt.barh(x,y,color='g')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('horizontal bar graph')
plt.show()
```

#Bubble Graph

**Bubble graph is a variation of scatter plot where each datapoint has different color and size**

```python
x=[1,2,3,4,5]
y=[3,5,6,10,5]
plt.scatter(x,y,s=np.random.rand(5)*1000,c=np.random.random((5)))
plt.show()
```

# Pie Chart

```python
data=[3,5,7,4,3]
ex=(0.1,0,0,0,0)
lb=['a','b','c','d','e']              #labels of each part
plt.pie(data,labels=lb,explode=ex,autopct='%1.1f%%')    #explode is to highlight a part of the pie chart  autopct is for percentage of each part 
plt.show()
```
